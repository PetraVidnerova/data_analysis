```python
import pandas as pd
```

#  Read data files and concat them 

```python
students_mat = pd.read_csv('student-mat.csv')
students_por = pd.read_csv('student-por.csv')
assert all(students_mat.columns == students_por.columns)

students = pd.concat([students_mat, students_por])
students.reset_index(inplace=True, drop=True)

print(students.shape)
students.head()
```

#  Delete duplicate students

There are students who attended both classes.

```python
id_columns = list(students.columns) 
id_columns.remove('absences')
id_columns.remove('G1')
id_columns.remove('G2')
id_columns.remove('G3')

students[students.duplicated(subset=id_columns, keep=False)]
```

```python
students.drop_duplicates(subset=id_columns, inplace=True)
students.shape
```

```python
students.reset_index(inplace=True, drop=True)
students.tail()
```

# Get rid of unwanted columns

```python
students = students.drop("school", axis=1)
students = students.drop("guardian", axis=1)
students = students.drop("reason", axis=1)
students = students.drop(["failures", "schoolsup", "famsup"], axis=1)
students.head()
```

#  Add average alcohol consumption

```python
students['Aalc'] = (5*students['Dalc'] + 2*students['Walc'])/7
print(students.shape)
students.head()
```

# Get rid of yes/no values

```python
students.replace('yes', True, inplace=True)
students.replace('no', False, inplace=True)
```

# Save to file

```python
students.to_csv("students.csv")
```
